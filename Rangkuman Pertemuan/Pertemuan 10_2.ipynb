# Variabel Baru dari Data Asli

```python
import pandas as pd

data_transaksi = pd.DataFrame({
    'ID Transaksi': [1, 2, 3, 4, 5],
    'ID Pelanggan': [101, 102, 103, 104, 105],
    'Jumlah Pembelian': [250000, 150000, 300000, 200000, 350000],
    'Tanggal Transaksi': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05']
})

data_transaksi['Jumlah Pembelian'] = data_transaksi['Jumlah Pembelian'].apply(lambda x: f"Rp {x:,}")

data_transaksi['Tanggal Transaksi'] = pd.to_datetime(data_transaksi['Tanggal Transaksi'])

data_transaksi['Hari'] = data_transaksi['Tanggal Transaksi'].dt.day
data_transaksi['Bulan'] = data_transaksi['Tanggal Transaksi'].dt.month
data_transaksi['Tahun'] = data_transaksi['Tanggal Transaksi'].dt.year
print(data_transaksi)
```

# Encoding

### 1. Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

data = pd.DataFrame({
    'id' : [1, 2, 3, 4, 5],
    'Tingkat' : ['Rendah', 'Sedang', 'Tinggi', 'Rendah', 'Tinggi']
})

le = LabelEncoder()
data['Tingkat_encoded'] = le.fit_transform(data['Tingkat'])
print(data)
```

### 2. One-Hot Encoding

```python
data_kategori = pd.DataFrame({
    'ID Produk': [1, 2, 3],
    'Kategori': ['Elektronik', 'Pakaian', 'Makanan']
})
data_kategori_encoded = pd.get_dummies(data_kategori, columns=['Kategori'], prefix='Kategori')

print(data_kategori_encoded)
```

# Normalisasi dan Standarisasi

### 1. Normalisasi (Min-Max Scaling)

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

data = pd.DataFrame({
    'Nilai_Matematika' : [70, 85, 90],
    'Nilai_Fisika' : [75, 80, 92]
})

data_copy = data.copy()

scaler = MinMaxScaler()
data_copy[['Nilai Matematika', 'Nilai_Fisika']] = scaler.fit_transform(data_copy[['Nilai_Matematika', 'Nilai_Fisika']])

print("Original DataFrame:")
print(data)
print('\nScaled DataFrame:')
print(data_copy)
```

### 2. Standarisasi (Standarization)

```python
from sklearn.preprocessing import StandardScaler

data_copy2 = data.copy()

scaler2 = StandardScaler()

data_copy2[['Nilai_Matematika', 'Nilai_Fisika']] = scaler2.fit_transform(data_copy2[['Nilai_Matematika','Nilai_Fisika']])

print("Original DataFrame:")
print(data)
print("\nStandard Scaled DataFrame:")
print(data_copy2)
```

# Transformasi Fitur dengan teknik Binning

```python
data_umur = pd.DataFrame({
    'Nama' : ['Ali', 'Budi', 'Cici', 'Dina', 'Eko', 'Fahri', 'Grace'],
    'Umur' : [15, 30, 25, 70, 101, -1, 0.5]
})

data_umur['Kategori_Umur'] = pd.cut(data_umur['Umur'], bins=[0, 18, 60, 100], labels=['Anak', 'Dewasa', 'Lansia'])

print(data_umur)
```

# Ekstraksi Informasi Waktu dari Kolom Tanggal

```python
#contoh 1 : Hari Bulan Tahun

import pandas as pd

data_transaksi = pd.DataFrame({
    'ID_Transaksi': [1, 2, 3, 4],
    'Tanggal_Transaksi': ['2023-01-15', '2023-03-12', '2023-06-20', '2023-09-25']
})

# Konversi kolom Tanggal_Transaksi menjadi tipe datetime
data_transaksi['Tanggal_Transaksi'] = pd.to_datetime(data_transaksi['Tanggal_Transaksi'])

# Ekstraksi informasi hari, bulan, dan tahun
data_transaksi['Hari'] = data_transaksi['Tanggal_Transaksi'].dt.day
data_transaksi['Bulan'] = data_transaksi['Tanggal_Transaksi'].dt.month
data_transaksi['Tahun'] = data_transaksi['Tanggal_Transaksi'].dt.year

print(data_transaksi)
```

```python
# contoh 2 : Hari dalam minggu dan kuartal

data_transaksi['Hari_dlm_Minggu'] = data_transaksi['Tanggal_Transaksi'].dt.day_name()
data_transaksi['Kuartal'] = data_transaksi['Tanggal_Transaksi'].dt.quarter

print(data_transaksi)
```

```python
# contoh 3 : Menggabungkan Hari Bulan Tahun jadi satu kolom

import pandas as pd

# Data contoh dengan kolom Hari, Bulan, dan Tahun terpisah
data_waktu = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Hari': [15, 12, 20, 25],
    'Bulan': [1, 3, 6, 9],
    'Tahun': [2023, 2023, 2023, 2023]
})

# Menampilkan data asli
print("Data sebelum digabung:")
print(data_waktu)
```

```python
# SOLUSI 1 = Ubah Nama Kolom

# Mengganti nama kolom
data_waktu = data_waktu.rename(columns={'Tahun': 'year', 'Bulan': 'month', 'Hari': 'day'})

# Menggabungkan kolom menjadi satu kolom Tanggal
data_waktu['Tanggal'] = pd.to_datetime(data_waktu[['year', 'month', 'day']])

# Menghapus kolom yang tidak diperlukan
data_waktu = data_waktu.drop(columns=['year', 'month', 'day'])

# Menampilkan data hasil
print("\nData setelah digabung menjadi satu kolom Tanggal:")
print(data_waktu)
```

```python
# SOLUSI 2 = Gunakan pd.to_datetime() dengan parameter dayfirst

import pandas as pd

# Membuat data contoh dengan kolom Hari, Bulan, dan Tahun
data_waktu = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Hari': [15, 12, 20, 25],
    'Bulan': [1, 3, 6, 9],
    'Tahun': [2023, 2023, 2023, 2023]
})

print(data_waktu)
print(" ")

# Menggabungkan kolom dengan format string
data_waktu['Tanggal'] = pd.to_datetime(data_waktu['Tahun'].astype(str) + '-' +
                                       data_waktu['Bulan'].astype(str) + '-' +
                                       data_waktu['Hari'].astype(str))

# Menghapus kolom yang tidak diperlukan
data_waktu = data_waktu.drop(columns=['Hari', 'Bulan', 'Tahun'])

# Menampilkan data hasil
print("\nData setelah digabung menjadi satu kolom Tanggal:")
print("After Edit: ")
print(data_waktu)
```
