# Teknik Penanganan Missing Data (Missing Data Handling Techniques)

Missing data atau **data yang hilang** adalah kondisi di mana beberapa entri atau nilai dalam dataset tidak tersedia. Penanganan missing data sangat penting karena data yang hilang dapat menyebabkan bias dan mengurangi akurasi model analisis.

Secara umum ada 3: 
- Penghapusan, 
- Imputasi, 
- Placeholder

1. Penghapusan Missing Data (Deletion Methods)

Ada dua jenis utama metode penghapusan yang bisa dilakukan ketika data hilang, yaitu penghapusan baris (row deletion) dan penghapusan kolom (column deletion).

**Kelebihan**: Mudah dilakukan dan dapat menjaga konsistensi dataset.

**Kekurangan**: Menghapus data yang hilang dalam jumlah besar dapat menyebabkan hilangnya informasi penting.

a. Penghapusan Baris (Row Deletion) <p>
Metode ini dilakukan dengan menghapus baris yang memiliki nilai yang hilang. Teknik ini efektif jika jumlah missing data sedikit, tetapi jika banyak baris yang terpengaruh, informasi yang penting mungkin hilang.

```python
import pandas as pd

# Contoh data dengan missing values
data = pd.DataFrame({
    'A': [1, 2, None, 4, 5],
    'B': [5, None, None, 8, 10],
    'C': [10, 20, 30, None, 50]
})

# Menghapus baris dengan missing data
Row_deletion = data.dropna()

print("Data Awal: ")
print(data)

print("Pembersihan Baris Data: ")
print(Row_deletion)
```

b. Penghapusan Kolom (Column Deletion) <p>
Jika terdapat terlalu banyak data yang hilang dalam sebuah kolom, kita bisa menghapus seluruh kolom tersebut.

```python
print("Data Awal: ")
print(data)

print("Pembersihan Kolom Data: ")

# Menghapus kolom dengan missing data
Column_deletion = data.dropna(axis=1)
print(Column_deletion)
```

2. Imputasi Data (Data Imputation Methods)
Imputasi data adalah proses menggantikan nilai yang hilang dengan nilai estimasi atau substitusi. Ada beberapa metode imputasi yang umum digunakan:

a. Mean/Median/Mode Imputation
Teknik ini menggantikan nilai yang hilang dengan nilai rata-rata (mean), median, atau modus (mode) dari kolom yang bersangkutan.

**Kelebihan**: Mudah diimplementasikan dan dapat menjaga ukuran dataset tetap utuh. <p>
**Kekurangan**: Dapat mengubah distribusi data asli dan tidak mempertimbangkan hubungan antar variabel.

- Imputasi Mean (Rata-Rata): Menggunakan rata-rata kolom untuk menggantikan missing values.

- Imputasi Median: Menggantikan nilai yang hilang dengan median (nilai tengah setelah data diurutkan).

- Imputasi Mode: Menggantikan nilai yang hilang dengan nilai yang paling sering muncul.

```python
print(data)

# Menggantikan nilai yang hilang dengan mean
mean_imputed_data = data.fillna(data.mean())
print(mean_imputed_data)

# Menggantikan nilai yang hilang dengan median
median_imputed_data = data.fillna(data.median())
print(median_imputed_data)

# Menggantikan nilai yang hilang dengan mode
mode_imputed_data = data.fillna(data.mode().iloc[0])
print(mode_imputed_data)
```

b. Forward Fill / Backward Fill <p>
Metode ini menggantikan nilai yang hilang dengan nilai sebelumnya (forward fill) atau nilai berikutnya (backward fill).

**Kelebihan**: Berguna dalam data time series dan data berurutan. <p>
**Kekurangan**: Mengasumsikan pola berurutan dalam data, yang mungkin tidak selalu sesuai.

- Forward Fill: Mengisi missing values dengan nilai dari entri sebelumnya.

```python
print(data)

# Menggunakan forward fill
ff_imputed_data = data.ffill()
print(ff_imputed_data)
```

- Backward Fill: Mengisi missing values dengan nilai dari entri berikutnya.

```python
print(data)

# Menggunakan backward fill
bf_imputed_data = data.bfill()

print(bf_imputed_data)
```

c. K-Nearest Neighbors Imputation (KNN Imputation) <p>
Metode ini menggunakan nilai dari beberapa tetangga terdekat (nearest neighbors) untuk mengisi nilai yang hilang.

Proses: Untuk setiap missing value, algoritma menghitung jarak (misalnya Euclidean distance) antara data yang hilang dengan data lain, kemudian nilai yang hilang diisi dengan rata-rata tetangga terdekat.

**Kelebihan**: Mempertimbangkan hubungan antar fitur. <p>
**Kekurangan**: Memerlukan lebih banyak komputasi dan bisa menjadi lambat untuk dataset besar.

```python
print(data)

#pip install scikit-learn
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=2)
knn_imputed_data = pd.DataFrame(imputer.fit_transform(data), columns=data.columns)

print(knn_imputed_data)
```

d. Multiple Imputation <p>
Imputasi berganda adalah metode di mana nilai yang hilang diisi beberapa kali untuk menghasilkan beberapa dataset yang lengkap. Setiap dataset dianalisis secara terpisah, dan hasil akhirnya dirata-rata untuk mendapatkan estimasi yang lebih baik.

Proses: Menghasilkan beberapa dataset dengan mengimputasi nilai yang hilang menggunakan distribusi yang sesuai, kemudian merata-ratakan hasil analisis dari semua dataset tersebut.

**Kelebihan**: Lebih akurat karena mempertimbangkan ketidakpastian dalam data yang hilang. <p>
**Kekurangan**: Prosesnya lebih kompleks dan memerlukan lebih banyak sumber daya komputasi.



```python
print(data)

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

imputer = IterativeImputer()
multi_imputed_data = pd.DataFrame(imputer.fit_transform(data), columns=data.columns)

print(multi_imputed_data)
```

3. Penggunaan Placeholder <p>
Menggunakan nilai placeholder untuk menggantikan missing data bisa menjadi solusi sementara ketika ingin mengidentifikasi dengan jelas bagian mana yang hilang tanpa secara langsung menghapus atau mengimputasinya.

**Kelebihan**: Dapat digunakan untuk penandaan atau pelacakan missing values. <p>
**Kekurangan**: Placeholder perlu ditangani dengan hati-hati saat melakukan analisis statistik, karena bisa mengganggu hasil perhitungan.

```python
print(data)

# Mengganti nilai yang hilang dengan placeholder (-999)
placeholder_data = data.fillna(-999)
print(placeholder_data)
```

Penanganan Missing Data pada Data Kategorik

Data kategorik memiliki sifat yang berbeda dibandingkan dengan data numerik, sehingga metode penanganan missing data pada data kategorik juga sedikit berbeda. Dalam dataset, missing data pada kolom dengan nilai kategorik harus ditangani secara hati-hati karena mengganti nilai yang hilang dengan cara yang tidak tepat dapat mengubah makna dan distribusi data tersebut.

Berikut adalah beberapa metode yang umum digunakan untuk menangani missing data pada kolom kategorik:

1. Menggantikan dengan Modus (Mode Imputation) <p>
Modus adalah nilai yang paling sering muncul dalam suatu kolom. Jika suatu kolom kategorik memiliki missing values, metode sederhana yang dapat digunakan adalah menggantinya dengan modus dari kolom tersebut.

Proses: Cari nilai yang paling sering muncul dalam kolom, lalu gantikan semua missing values dengan nilai tersebut.

**Kelebihan**: Mudah dilakukan dan memastikan data tetap konsisten. <p>
**Kekurangan**: Jika missing values terlalu banyak, penggantian dengan modus dapat menyebabkan dominasi satu kategori yang berlebihan, sehingga mengurangi variasi.

```python
import pandas as pd

# Contoh data kategorik dengan missing values
data_kategorik = pd.DataFrame({
    'Produk': ['Laptop', 'Smartphone', 'Tablet', None, 'Smartphone', 'Tablet', None],
    'Brand': ['Apple', 'Samsung', 'Apple', 'Samsung', None, 'Apple', 'Samsung']
})
print(data_kategorik)
print(" ")

# Mengganti missing values dengan modus
mode_imputed_data = data_kategorik.fillna(data_kategorik.mode().iloc[0])
print(mode_imputed_data)
```

2. Penggantian dengan Kategori Baru (New Category Imputation) <p>
Salah satu cara paling sederhana untuk menangani missing data dalam kolom kategorik adalah dengan membuat kategori baru yang mewakili nilai yang hilang, misalnya dengan label seperti “Unknown” atau “Missing”.

Proses: Tambahkan kategori baru (misalnya "Unknown") untuk menggantikan semua missing values.

**Kelebihan**: Tidak mempengaruhi distribusi asli dari kategori yang ada. <p>
**Kekurangan**: Kategori “Unknown” mungkin tidak memiliki makna yang relevan dan bisa dianggap bias jika terdapat banyak missing values.

```python
print(data_kategorik)
print(" ")

# Menggantikan missing values dengan kategori baru 'Unknown'
data_kategorik_imputed = data_kategorik.fillna('Unknown')
print(data_kategorik_imputed)
```

3. K-Nearest Neighbors Imputation untuk Data Kategorik <p>
Sama seperti pada data numerik, K-Nearest Neighbors (KNN) juga dapat digunakan untuk mengimputasi data kategorik. Algoritma ini menggantikan nilai yang hilang berdasarkan nilai-nilai yang paling dekat (nearest neighbors) dengan data yang lengkap.

Proses: KNN menghitung kesamaan atau jarak antara entri-entri yang hilang dengan entri-entri lainnya dan mengganti missing values berdasarkan kategori dari tetangga terdekat.

**Kelebihan**: Mempertimbangkan hubungan antara fitur dalam dataset. <p>
**Kekurangan**: Proses lebih kompleks dan lambat untuk dataset yang besar.

```python
print(data_kategorik)
print(" ")

from sklearn.impute import KNNImputer
import pandas as pd

# Data kategorik harus dikodekan sebagai angka terlebih dahulu (misalnya dengan One-Hot Encoding)
data_encoded = pd.get_dummies(data_kategorik, drop_first=True)

# Gunakan KNN untuk imputasi
imputer = KNNImputer(n_neighbors=2)
knn_imputed_data = pd.DataFrame(imputer.fit_transform(data_encoded), columns=data_encoded.columns)

# Dekode kembali ke kategori asli jika diperlukan

print(knn_imputed_data)
```

4. Decision Tree Imputation <p>
Metode ini menggunakan algoritma decision tree untuk memprediksi kategori dari nilai yang hilang berdasarkan variabel lain dalam dataset.

Proses: Buat model prediksi dengan decision tree untuk mempelajari pola dari variabel lain dan prediksi nilai yang hilang pada variabel kategorik.

**Kelebihan**: Memanfaatkan informasi dari variabel lain. <p>
**Kekurangan**: Membutuhkan pengetahuan lebih lanjut tentang machine learning dan bisa memerlukan waktu untuk proses training.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Contoh data kategorik dengan missing values
data_kategorik = pd.DataFrame({
    'Produk': ['Laptop', 'Smartphone', 'Tablet', None, 'Smartphone', 'Tablet', None],
    'Brand': ['Apple', 'Samsung', 'Apple', 'Samsung', None, 'Apple', 'Samsung']
})

# Menghapus baris yang memiliki missing values pada kolom 'Brand'
data_kategorik = data_kategorik.dropna(subset=['Brand'])

# Menghilangkan missing values untuk pelatihan decision tree
data_kategorik_clean = data_kategorik.dropna().copy()  # Gunakan .copy() untuk menghindari peringatan

# Label encoding untuk mengubah kategori menjadi angka
le_brand = LabelEncoder()
le_produk = LabelEncoder()

# Transformasi data kategorik menjadi numerik
data_kategorik_clean['Brand_encoded'] = le_brand.fit_transform(data_kategorik_clean['Brand'])
data_kategorik_clean['Produk_encoded'] = le_produk.fit_transform(data_kategorik_clean['Produk'])

# Membuat model Decision Tree Classifier
model = DecisionTreeClassifier()
X = data_kategorik_clean[['Brand_encoded']]  # Fitur (Brand yang sudah diencode)
y = data_kategorik_clean['Produk_encoded']   # Target yang akan diprediksi (Produk yang sudah diencode)
model.fit(X, y)

# Salinan data_kategorik untuk hasil prediksi
predicted_data_kategorik = data_kategorik.copy()

# Transformasi kolom 'Brand' yang hilang ke dalam bentuk numerik untuk prediksi
missing_data = predicted_data_kategorik[predicted_data_kategorik['Produk'].isna()].copy()  # Salin agar menghindari peringatan
missing_data['Brand_encoded'] = le_brand.transform(missing_data['Brand'])

# Prediksi missing values pada kolom 'Produk' (berdasarkan Brand)
predicted_produk = model.predict(missing_data[['Brand_encoded']])

# Decode hasil prediksi kembali ke kategori aslinya
predicted_data_kategorik.loc[predicted_data_kategorik['Produk'].isna(), 'Produk'] = le_produk.inverse_transform(predicted_produk)

# Menampilkan hasil prediksi dalam variabel baru tanpa overwrite data_kategorik asli
print("Data asli (data_kategorik):")
print(data_kategorik)

print("\nHasil prediksi dalam variabel baru (predicted_data_kategorik):")
print(predicted_data_kategorik)
```

Jenis Missing Data dan Strategi Penanganan:

1. Missing Completely at Random (MCAR): Data hilang secara acak, dan tidak terkait dengan variabel lain dalam dataset.

**Strategi**: Imputasi atau penghapusan baris/kolom.

2. Missing at Random (MAR): Data hilang terkait dengan satu atau lebih variabel lainnya, tetapi tidak dengan variabel yang hilang itu sendiri.

**Strategi**: Imputasi atau metode canggih seperti multiple imputation.

3. Missing Not at Random (MNAR): Data hilang tidak secara acak, dan hilangnya data terkait langsung dengan variabel yang hilang itu sendiri.

**Strategi**: Lebih kompleks, bisa menggunakan model statistik untuk memperkirakan data yang hilang.

# Teknik Data Cleaning: Pembersihan Data

Data cleaning adalah proses penting dalam analisis data yang bertujuan untuk memastikan bahwa data yang digunakan bersih, konsisten, dan bebas dari kesalahan. Data yang kotor dapat menyebabkan bias, kesalahan dalam analisis, dan hasil yang tidak valid. Dalam proses pembersihan data, berbagai teknik digunakan untuk mengidentifikasi, menangani, dan memperbaiki data yang tidak valid atau tidak konsisten.



Berikut adalah beberapa teknik umum dalam data cleaning:

1. Menghapus Data Duplikat (Handling Duplicates) <p>
Data duplikat terjadi ketika dua atau lebih baris di dataset memiliki nilai yang sama di seluruh kolom atau di beberapa kolom kunci tertentu. Data duplikat bisa muncul akibat proses input data yang tidak benar atau penggabungan data dari berbagai sumber.

```python
import pandas as pd

# Contoh data dengan duplikasi
data = pd.DataFrame({
    'ID': [1, 2, 3, 4, 2],
    'Nama': ['Alice', 'Bob', 'Charlie', 'David', 'Bob'],
    'Umur': [25, 30, 35, 40, 30]
})

# Menampilkan data duplikat
print("Data sebelum pembersihan duplikasi:")
print(data)

# Menghapus baris duplikat
data_cleaned = data.drop_duplicates()

print("\nData setelah pembersihan duplikasi:")
print(data_cleaned)
```

3. Menangani Format Data yang Tidak Konsisten (Inconsistent Data Formatting) <p>
Sering kali, format data di dalam satu kolom bisa tidak konsisten, misalnya nama yang diawali dengan huruf besar atau kecil, tanggal dengan format berbeda, atau data numerik yang disimpan sebagai string. Konsistensi format sangat penting untuk memastikan keakuratan analisis.

```python
import pandas as pd

# Contoh data dengan format yang tidak konsisten
data_format = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Tanggal': ['2022-01-01', '01/02/2022', 'March 3, 2022', '2022-04-01'],
    'Pendapatan': ['10000', '$20000', '30000', '40,000']
})

# Menampilkan data asli
print("Data sebelum perbaikan format:")
print(data_format)

# Mengubah kolom 'Pendapatan' menjadi tipe numerik
data_format['Pendapatan'] = data_format['Pendapatan'].replace({'\$': '', ',': ''}, regex=True).astype(float)

# Menstandarkan format tanggal
data_format['Tanggal'] = pd.to_datetime(data_format['Tanggal'])

# Menampilkan data setelah perbaikan
print("\nData setelah perbaikan format:")
print(data_format)
```

4. Menangani Nilai yang Salah (Incorrect Data Values) <p>
Nilai yang salah bisa muncul akibat kesalahan input, misalnya nilai umur negatif atau nilai yang tidak relevan dengan konteks. Dalam hal ini, perlu dilakukan validasi data dan penggantian nilai yang salah.

```python
# Contoh data dengan nilai yang salah
data_invalid = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Nama': ['Alice', 'Bob', 'Charlie', 'David'],
    'Umur': [25, -30, 35, 200]  # Nilai negatif dan umur yang tidak logis
})

# Menampilkan data asli
print("Data sebelum pembersihan nilai yang salah:")
print(data_invalid)

# Mengganti nilai yang salah dengan NaN
data_invalid['Umur'] = data_invalid['Umur'].apply(lambda x: x if 0 < x < 120 else None)

print("\nData setelah pembersihan nilai yang salah:")
print(data_invalid)
```
