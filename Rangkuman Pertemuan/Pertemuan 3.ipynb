```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#create a sample dataframe
data = {
    "Product": ["A", "B", "A", "B", "C", "C", "A"],
    "Category": ["Electronics", "Clothing", "Electronics", "Clothing", "Home", "Home", "Electronics"],
    "Sales": [100, 150, 200, 250, 300, 350, 400],
    "Quantity": [10, 15, 20, 25, 30, 35, 40]
}
df = pd.DataFrame(data)
print(df)
print(df.dtypes)

print("\n")

#information about the dataframe
df.info()

print("\n")

#convert Sales to float
df["Sales"] = df["Sales"].astype(float)
print(df.dtypes)

print("\n")

#group by Category and calculate total Sales and average Quantity
summary = df.groupby("Category").agg({
    "Sales": "sum",
    "Quantity": "mean"
})
print(summary)

print("\n")

#group by Product and calculate total Sales and average Quantity
grouped = df.groupby("Product").agg({
    "Sales": "sum",
    "Quantity": "mean"
})
print(grouped)

print("\n")

#Selecting Product and Sales Columns
selected_columns = df[["Product", "Sales"]]
print(selected_columns)

print("\n")

#add a new column for Revenue (Sales * Quantity)
df['Revenue'] = None
print(df)

print("\n")

#Sorting by sales in descending order
sorted_df = df.sort_values(by="Sales", ascending=False)
print(sorted_df)

print("\n")

#Sorting by sales and then by Sales in ascending order
sorted_df = df.sort_values(by="Sales", ascending=True)
print(sorted_df)

print("\n")

#contoh 1
print("Contoh 1: Pie Chart")
Labels = ['A', 'B', 'C']
Values = [700, 400, 650]

plt.pie(Values, labels=Labels, autopct='%1.1f%%', startangle= 140)
plt.axis('equal')
plt.title('Sales Distribution by Product')
plt.show()

#contoh 2
print("Contoh 2: Bar Chart")
categories = ['Sangat Baik','Baik', 'Cukup Baik', 'Kurang Baik', 'Tidak Baik']
values = [50, 30, 15, 4, 1]

plt.bar(categories, values, color=['green', 'blue', 'yellow', 'orange', 'red'])
plt.xlabel('Kualitas Pelayanan')
plt.ylabel('Jumlah Responden')
plt.title('Distribusi Kualitas Pelayanan')  
plt.show()

#contoh 3
print("Contoh 3: Histogram")
np.random.seed(0)
temperature = np.random.normal(loc=20, scale=5, size=1000)
plt.hist(temperature, bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')
plt.title('Distribution of Daily Temperatures')
plt.show()

#contoh 4
print("Contoh 4: Scatter Plot")
height = np.random.normal(loc=170, scale=10, size=1000)
weight = np.random.normal(loc=70, scale=15, size=1000)

plt.scatter(height, weight, alpha=0.5)
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.title('Height vs Weight')
plt.show()

# Pie chart berdasarkan 'Product' dan 'Sales'
plt.figure(figsize=(7, 7))
plt.pie(sorted_df['Sales'], labels=sorted_df['Product'], autopct='%1.1f%%', startangle=90)
plt.title('Sales Distribution by Product')
plt.savefig('sales_analysis1.png')
plt.show()


# Bar plot untuk 'Product' dan 'Sales'
plt.figure(figsize=(8, 6))
plt.bar(sorted_df['Product'], sorted_df['Sales'], color='skyblue')
plt.title('Sales by Product')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.savefig('sales_analysis2.png')
plt.show()


# Histogram untuk distribusi 'Sales'
plt.figure(figsize=(8, 6))
plt.hist(sorted_df['Sales'], bins=5, color='green', edgecolor='black')
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.savefig('sales_analysis3.png')
plt.show()


# Scatterplot untuk 'Sales' dan 'Quantity'
plt.figure(figsize=(8, 6))
plt.scatter(sorted_df['Sales'], sorted_df['Quantity'], color='red', s=100)
plt.title('Sales vs Quantity')
plt.xlabel('Sales')
plt.ylabel('Quantity')
plt.savefig('sales_analysis4.png')
plt.show()

```
