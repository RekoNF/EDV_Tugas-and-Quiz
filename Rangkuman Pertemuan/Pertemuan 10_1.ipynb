# 1. Pembuatan Variabel Baru dari Data Asli

Sering kali, data mentah memiliki variabel yang sebenarnya dapat memberikan lebih banyak informasi jika diubah atau diolah dengan cara tertentu. Misalnya, data penjualan mungkin memiliki kolom tanggal, namun hanya berupa tanggal transaksi saja. Dari kolom ini, kita bisa membuat kolom baru seperti **Hari**, **Bulan**, dan **Tahun** untuk menangkap informasi waktu lebih spesifik.

Contoh: Ekstraksi Informasi dari Tanggal

Misalnya kita memiliki data transaksi dengan kolom Tanggal_Transaksi, dan kita ingin membuat variabel baru **Hari**, **Bulan**, dan **Tahun**.

```python
import pandas as pd

# Data transaksi
data_transaksi = pd.DataFrame({
    'ID_Transaksi': [1, 2, 3, 4],
    'Tanggal_Transaksi': ['2023-01-15', '2023-03-12', '2023-06-20', '2023-09-25']
})

# Konversi kolom Tanggal_Transaksi menjadi tipe datetime
data_transaksi['Tanggal_Transaksi'] = pd.to_datetime(data_transaksi['Tanggal_Transaksi'])

# Membuat fitur baru
data_transaksi['Hari'] = data_transaksi['Tanggal_Transaksi'].dt.day
data_transaksi['Bulan'] = data_transaksi['Tanggal_Transaksi'].dt.month
data_transaksi['Tahun'] = data_transaksi['Tanggal_Transaksi'].dt.year

print(data_transaksi)
```

Variabel baru **Hari**, **Bulan**, dan **Tahun** memberikan detail tambahan yang lebih kaya bagi model.

# 2. Encoding Variabel Kategorikal
Sebagian besar model machine learning memerlukan data dalam bentuk numerik, sehingga kita perlu mengonversi variabel kategorikal menjadi bentuk numerik. Teknik encoding bertujuan untuk memastikan model dapat memanfaatkan informasi kategori dengan benar. Ada beberapa teknik encoding umum:

1. Label Encoding

*Label Encoding* mengubah setiap kategori dalam fitur menjadi nilai numerik yang unik. Teknik ini sesuai jika variabel memiliki urutan alami, misalnya kategori low, medium, dan high.

**Contoh Label Encoding**

```python
from sklearn.preprocessing import LabelEncoder

# Data kategori
data = pd.DataFrame({
    'ID': [1, 2, 3],
    'Tingkat': ['Rendah', 'Sedang', 'Tinggi']
})

# Inisialisasi LabelEncoder
le = LabelEncoder()
data['Tingkat_Encoded'] = le.fit_transform(data['Tingkat'])

print(data)
```

2. One-Hot Encoding

*One-Hot Encoding* mengubah setiap kategori dalam fitur menjadi kolom biner. Ini efektif untuk variabel tanpa urutan tertentu.

**Contoh One-Hot Encoding**

```python
data_kategori = pd.DataFrame({
    'ID': [1, 2, 3],
    'Jenis_Produk': ['Laptop', 'Tablet', 'Smartphone']
})

# One-Hot Encoding
data_kategori_encoded = pd.get_dummies(data_kategori, columns=['Jenis_Produk'], prefix='Produk')

print(data_kategori_encoded)
```

Kolom **Produk_Laptop**, **Produk_Tablet**, dan **Produk_Smartphone** sekarang berupa kolom biner yang merepresentasikan kategori awal.

# 3. Normalisasi dan Standarisasi

Normalisasi dan standarisasi mengubah skala fitur agar memiliki nilai yang seragam. Ini penting karena data yang memiliki skala terlalu besar atau terlalu kecil bisa memperlambat pelatihan model dan menurunkan performanya.

1. Normalisasi (Min-Max Scaling)

Normalisasi mengubah nilai data dalam suatu kolom ke rentang tertentu, biasanya antara 0 dan 1.

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# Create the DataFrame
data = pd.DataFrame({
    'Nilai_Matematika': [70, 85, 90],
    'Nilai_Fisika': [75, 80, 92]
})

# Properly create a copy of the DataFrame
data_copy = data.copy()

# Apply MinMaxScaler to the copied DataFrame
scaler = MinMaxScaler()
data_copy[['Nilai_Matematika', 'Nilai_Fisika']] = scaler.fit_transform(data_copy[['Nilai_Matematika', 'Nilai_Fisika']])

# Display the original and scaled DataFrames
print("Original DataFrame:")
print(data)
print("\nScaled DataFrame:")
print(data_copy)
```

2. Standarisasi (Standardization)

Standarisasi mengubah nilai data agar memiliki rata-rata 0 dan standar deviasi 1.

```python
from sklearn.preprocessing import StandardScaler

# Create another copy of the DataFrame for StandardScaler
data_copy2 = data.copy()

# Apply StandardScaler to the copied DataFrame
scaler = StandardScaler()
data_copy2[['Nilai_Matematika', 'Nilai_Fisika']] = scaler.fit_transform(data_copy2[['Nilai_Matematika', 'Nilai_Fisika']])

# Display the original and StandardScaler-scaled DataFrames
print("Original DataFrame:")
print(data)
print("\nStandard Scaled DataFrame:")
print(data_copy2)
```

# 4, Transformasi Fitur dengan Teknik Binning

*Binning* atau **discretization** adalah teknik mengelompokkan nilai numerik menjadi kategori. Contohnya, data usia bisa diubah menjadi kategori **Anak**, **Dewasa**, dan **Lansia**.

**Contoh Binning**

```python
data_umur = pd.DataFrame({
    'Nama': ['Ali', 'Budi', 'Cici', 'Dina'],
    'Umur': [15, 30, 25, 70]
})

data_umur['Kategori_Umur'] = pd.cut(data_umur['Umur'], bins=[0, 18, 60, 100], labels=['Anak', 'Dewasa', 'Lansia'])

print(data_umur)
```

# 5. Ekstraksi Informasi Waktu dari Kolom Tanggal

Dalam **feature engineering**, kita sering kali perlu mengekstraksi informasi yang lebih spesifik dari kolom tanggal atau waktu. Informasi ini dapat berupa hari, bulan, tahun, atau bahkan hari dalam minggu dan kuartal. Dengan mengekstraksi informasi waktu ini, kita bisa membantu model menemukan pola berbasis waktu yang mungkin ada di dalam data. Beberapa contoh umum termasuk pola musiman, pola mingguan, dan pola harian yang relevan dalam analisis penjualan, perilaku pengguna, atau prediksi cuaca.

Contoh 1: Ekstraksi Hari, Bulan, dan Tahun

Misalkan kita memiliki data transaksi yang hanya menyertakan kolom Tanggal_Transaksi. Untuk membantu analisis berbasis waktu, kita dapat menambahkan kolom yang memisahkan informasi menjadi Hari, Bulan, dan Tahun.

```python
import pandas as pd

# Data transaksi
data_transaksi = pd.DataFrame({
    'ID_Transaksi': [1, 2, 3, 4],
    'Tanggal_Transaksi': ['2023-01-15', '2023-03-12', '2023-06-20', '2023-09-25']
})

# Konversi kolom Tanggal_Transaksi menjadi tipe datetime
data_transaksi['Tanggal_Transaksi'] = pd.to_datetime(data_transaksi['Tanggal_Transaksi'])

# Ekstraksi informasi hari, bulan, dan tahun
data_transaksi['Hari'] = data_transaksi['Tanggal_Transaksi'].dt.day
data_transaksi['Bulan'] = data_transaksi['Tanggal_Transaksi'].dt.month
data_transaksi['Tahun'] = data_transaksi['Tanggal_Transaksi'].dt.year

print(data_transaksi)
```

Contoh 2: Ekstraksi Hari dalam Minggu dan Kuartal

Selain informasi dasar seperti hari, bulan, dan tahun, kita juga bisa mengekstrak hari dalam minggu (misalnya, apakah hari tersebut Senin atau Jumat) dan kuartal (apakah kuartal pertama, kedua, dan seterusnya). Informasi ini berguna ketika data memiliki pola mingguan atau kuartalan.

```python
# Ekstraksi hari dalam minggu dan kuartal
data_transaksi['Hari_dlm_Minggu'] = data_transaksi['Tanggal_Transaksi'].dt.day_name()
data_transaksi['Kuartal'] = data_transaksi['Tanggal_Transaksi'].dt.quarter

print(data_transaksi)
```

Dalam contoh ini:

- Hari_dlm_Minggu menunjukkan nama hari (misalnya, Sunday, Tuesday).
- Kuartal menunjukkan kuartal tahun transaksi dilakukan (1, 2, 3, atau 4).

Jika data memiliki kolom yang terpisah untuk hari, bulan, dan tahun, kita bisa menggabungkannya menjadi satu kolom dengan format tanggal standar, seperti "year-month-day". Dengan menggabungkan informasi ini ke dalam satu variabel tanggal, kita akan memiliki format yang lebih konsisten dan lebih mudah untuk dianalisis, terutama dalam analisis berbasis waktu atau saat menggunakan model prediktif yang memerlukan variabel waktu.

Berikut adalah contoh cara menggabungkan kolom Hari, Bulan, dan Tahun menjadi satu kolom tanggal dengan format year-month-day menggunakan Python.

Contoh 3: Menggabungkan Hari, Bulan, dan Tahun menjadi Satu Kolom Tanggal

```python
import pandas as pd

# Data contoh dengan kolom Hari, Bulan, dan Tahun terpisah
data_waktu = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Hari': [15, 12, 20, 25],
    'Bulan': [1, 3, 6, 9],
    'Tahun': [2023, 2023, 2023, 2023]
})

# Menampilkan data asli
print("Data sebelum digabung:")
print(data_waktu)
```

Solusi 1: Ubah Nama Kolom

Ganti nama kolom Tahun, Bulan, dan Hari menjadi year, month, dan day sebelum mengonversi:

```python
# Mengganti nama kolom
data_waktu = data_waktu.rename(columns={'Tahun': 'year', 'Bulan': 'month', 'Hari': 'day'})

# Menggabungkan kolom menjadi satu kolom Tanggal
data_waktu['Tanggal'] = pd.to_datetime(data_waktu[['year', 'month', 'day']])

# Menghapus kolom yang tidak diperlukan
data_waktu = data_waktu.drop(columns=['year', 'month', 'day'])

# Menampilkan data hasil
print("\nData setelah digabung menjadi satu kolom Tanggal:")
print(data_waktu)
```

Solusi 2: Gunakan pd.to_datetime() dengan Parameter dayfirst

Anda juga bisa menggabungkan kolom secara manual menggunakan pd.to_datetime() dengan format string:

```python
import pandas as pd

# Membuat data contoh dengan kolom Hari, Bulan, dan Tahun
data_waktu = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Hari': [15, 12, 20, 25],
    'Bulan': [1, 3, 6, 9],
    'Tahun': [2023, 2023, 2023, 2023]
})

print(data_waktu)
print(" ")

# Menggabungkan kolom dengan format string
data_waktu['Tanggal'] = pd.to_datetime(data_waktu['Tahun'].astype(str) + '-' +
                                       data_waktu['Bulan'].astype(str) + '-' +
                                       data_waktu['Hari'].astype(str))

# Menghapus kolom yang tidak diperlukan
data_waktu = data_waktu.drop(columns=['Hari', 'Bulan', 'Tahun'])

# Menampilkan data hasil
print("\nData setelah digabung menjadi satu kolom Tanggal:")
print("After Edit: ")
print(data_waktu)
```
