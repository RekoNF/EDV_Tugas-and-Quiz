## Import Dataset

Sumber dataset : [Kaggle](https://www.kaggle.com/datasets/shree0910/ai-and-data-science-job-market-dataset-20202026)

The AI & Data Science Job Market Dataset (2020–2026) is a synthetically generated dataset designed to simulate real-world hiring patterns across the artificial intelligence and data science job market.

The dataset contains structured information about job roles, company characteristics, required technical skills, education levels, experience requirements, and salary ranges. It reflects hiring data across multiple countries, industries, and company sizes.

This dataset is designed to support data analysis, machine learning model development, and business intelligence projects.

The dataset can be used for:

- Salary prediction models
- Job demand analysis
- Machine learning classification tasks
- Workforce analytics
- Skill demand trend analysis
- Data science portfolio projects
- All data in this dataset is synthetically generated using statistical distributions and controlled randomness, ensuring privacy while maintaining realistic patterns

### Import and Preview Data

```python
#import data sets

data = pd.read_csv(r"C:\Users\Mareko\OneDrive\Documents\Python Statistika\DataSets\AI Job Market Dataset.csv", delimiter=",")
df = pd.DataFrame(data)
df = df.head(101)
print(df)
```

Menghilangkan Kolom yang tidak perlu

```python
df = df.drop(['job_id','skills_ml', 'skills_deep_learning', 'skills_cloud', 'job_posting_month', 'job_posting_year'], axis=1)
print(df)
```

Melakukan filter tipe data

```python
df = df.dtypes
print(df)
```

## Statistik Deskriptif

```python
# Statistik Deskriptif
# Pastikan df adalah DataFrame yang memiliki kolom 'job_title' dan 'salary'
df = pd.DataFrame(data)
summary = df.groupby('job_title').agg({'salary': 'mean'})
print(summary)
summary1 = df.groupby('job_title').agg({'years_experience': 'mean'})
print(summary1) 
```

## Visualisasi Heatmap dan Pairplot

### Heatmap

```python
print(df.select_dtypes(include='number').columns)

df2 = df.head(10)

plt.figure(figsize=(10,8))
sns.heatmap(df2[['years_experience', 'skills_python', 'skills_sql',
       'skills_ml', 'skills_deep_learning', 'skills_cloud',
       'job_posting_month', 'job_openings']], annot=True, cmap='YlGnBu')
plt.title("Visualisasi Array 2D (Sample)")
plt.show()
```

### Pairplot

```python
plt.figure(figsize=(20,12))
sns.pairplot(df[['years_experience', 'salary']])
plt.suptitle("Pairplot untuk Variabel Numerik", y=1.05)
plt.show()
```
