## Import Modul

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
```

## Import Data 

```python
data = pd.read_csv(r"C:\Users\Mareko\OneDrive\Documents\Python Statistika\DataSets\nvidia_stock_data_1999_2026.csv")
dframe = pd.DataFrame(data)
dframe = dframe.fillna(0)
#semua data yang kosong diisi dengan 0
print(dframe.head(51))
```

## Lakukan ANOVA

### Contoh 1

```python
model = ols('close ~ open', data=dframe).fit()
anova_results = anova_lm(model)
print(anova_results)
```

### Contoh 2

```python
# Mendapatkan p-value untuk faktor 'open', 'high', dan 'low' pada regresi linear berganda saham NVIDIA
model = ols('close ~ open + high + low', data=dframe).fit()
anova_results = anova_lm(model)
print(anova_results)

# Mendapatkan p-value untuk masing-masing variabel
p_value_open = anova_results.loc['open', 'PR(>F)'] if 'open' in anova_results.index else None
p_value_high = anova_results.loc['high', 'PR(>F)'] if 'high' in anova_results.index else None
p_value_low = anova_results.loc['low', 'PR(>F)'] if 'low' in anova_results.index else None

alpha = 0.05 # Tingkat signifikansi 5%

# Logika if-else berdasarkan p-value untuk open
if p_value_open is not None and p_value_open < alpha:
    print(f"p-value untuk open: {p_value_open:.5f} < {alpha}")
    print("Hasil: Tolak H0. Ada pengaruh signifikan dari open terhadap close.")
elif p_value_open is not None:
    print(f"p-value untuk open: {p_value_open:.5f} >= {alpha}")
    print("Hasil: Gagal menolak H0. Tidak ada pengaruh signifikan dari open terhadap close.")

# Logika if-else berdasarkan p-value untuk high
if p_value_high is not None and p_value_high < alpha:
    print(f"p-value untuk high: {p_value_high:.5f} < {alpha}")
    print("Hasil: Tolak H0. Ada pengaruh signifikan dari high terhadap close.")
elif p_value_high is not None:
    print(f"p-value untuk high: {p_value_high:.5f} >= {alpha}")
    print("Hasil: Gagal menolak H0. Tidak ada pengaruh signifikan dari high terhadap close.")

# Logika if-else berdasarkan p-value untuk low
if p_value_low is not None and p_value_low < alpha:
    print(f"p-value untuk low: {p_value_low:.5f} < {alpha}")
    print("Hasil: Tolak H0. Ada pengaruh signifikan dari low terhadap close.")
elif p_value_low is not None:
    print(f"p-value untuk low: {p_value_low:.5f} >= {alpha}")
    print("Hasil: Gagal menolak H0. Tidak ada pengaruh signifikan dari low terhadap close.")
```

## Model Regresi Linear dan Penjelasannya

### Contoh Model Regresi Linear Berganda pada Data Saham NVIDIA

```python
# Model regresi linear berganda: memprediksi harga penutupan (close) berdasarkan open, high, dan low
import statsmodels.api as sm
X = dframe[['open', 'high', 'low']]
y = dframe['close']
X = sm.add_constant(X)  # menambahkan intercept
model = sm.OLS(y, X).fit()
print(model.summary())
```

#### Penjelasan Sederhana Model Regresi Linear Berganda
Model regresi linear berganda yang dibuat bertujuan untuk memprediksi harga penutupan saham NVIDIA (`close`) berdasarkan tiga variabel prediktor, yaitu harga pembukaan (`open`), harga tertinggi (`high`), dan harga terendah (`low`) pada hari yang sama.

Secara matematis, model ini dapat dituliskan sebagai:
$$\text{close} = \beta_0 + \beta_1 \times \text{open} + \beta_2 \times \text{high} + \beta_3 \times \text{low} + \epsilon$$

Keterangan:
- $\beta_0$ = intercept (konstanta)
- $\beta_1, \beta_2, \beta_3$ = koefisien regresi untuk masing-masing variabel
- $\epsilon$ = error/residual
 
Setiap koefisien menunjukkan seberapa besar pengaruh variabel tersebut terhadap harga penutupan, dengan asumsi variabel lain tetap. Jika koefisien positif, maka kenaikan variabel tersebut cenderung meningkatkan harga penutupan.

```python
print(f"Model Adj. R-squared: {model.rsquared_adj}")
print(f"Model R-squared: {model.rsquared}")
```

```python
# Untuk memeriksa confidence interval dari koefisien regresi
ci = model.conf_int(alpha=0.05)
ci.columns = ['Lower Bound', 'Upper Bound']
print(ci)
print("\nDetail Confidence Interval:")
if 'open' in ci.index:
    print(f"Confidence Interval untuk open: {ci.loc['open'].values}")
if 'high' in ci.index:
    print(f"Confidence Interval untuk high: {ci.loc['high'].values}")
if 'low' in ci.index:
    print(f"Confidence Interval untuk low: {ci.loc['low'].values}")
print("\nInterpretasi:")
if ci is not None:
    for var in ['open', 'high', 'low']:
        if var in ci.index:
            lower, upper = ci.loc[var]
            if lower > 0 or upper < 0:
                print(f"Koefisien untuk {var} signifikan (CI tidak mencakup 0).")
            else:
                print(f"Koefisien untuk {var} tidak signifikan (CI mencakup 0).")
```

## Adj. R-Squared Meaning

Adjusted $R^2$ adalah versi modifikasi dari koefisien determinasi ($R^2$) yang menyesuaikan jumlah prediktor (variabel independen) dalam model regresi. Berbeda dengan $R^2$ biasa yang selalu naik saat variabel ditambah, adjusted $R^2$ hanya meningkat jika variabel baru meningkatkan akurasi model, menjadikannya metrik lebih baik untuk regresi berganda.
