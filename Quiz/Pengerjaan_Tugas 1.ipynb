Menggunakan Data dari BPS : [Persentase Rumah Tangga Menurut Provinsi dan Fasilitas Tempat Buang Air Besar, 2021](https://www.bps.go.id/id/statistics-table/2/MTQ3IzI=/persentase-rumah-tangga-menurut-provinsi-dan-fasilitas-tempat-buang-air-besar.html)

## - Import Module

```python
import pandas as pd
import openpyxl as opxl
```

## - Import Data

```python
data_analysis = pd.read_csv(r"C:\Users\Mareko\Downloads\Persentase Rumah Tangga Menurut Provinsi dan Fasilitas Tempat Buang Air Besar, 2021.csv", sep=",", skiprows= 2, header=1)
data_analysis = data_analysis.replace("-","0")
```

## - Make Data Frame

```python
df = pd.DataFrame(data_analysis)
print(df)
```

## - Split data dengan iloc

```python
# Memilih baris data ke 15 / index ke 16
split1 = df.iloc[[15]]
print(split1)
print(" ")

print(" ")
# Memilih baris data ke 19 / index ke 20
split2 = df.iloc[[19]]
print(split2)

split1.index = split2.index
```

## - Merge data

```python
data_merge = pd.merge(split1, split2, left_index=True, right_index=True, suffixes=('_1','_2'))
print(data_merge)
```

## - Export output to excel

```python
data_merge.to_csv('hasil_merge.csv', sep=';', index=False)
print("\nHasil telah diexport ke 'hasil_merge.csv'")

data_merge.to_excel('hasil_merge.xlsx', index=False)
print("\nHasil telah diexport ke 'hasil_merge.xlsx'")
```
